```python
%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import re
import os
```

```python

def plot_df(df):
    _ = [df.__delitem__(col) for col in df.columns if '_relativ' in col]
    plt.style.use('ggplot')
    ax = df.plot(
        title='Korrelation der Begriffe {}'.format(" und ".join(df.columns.values)))
  
```

```python
df = pd.DataFrame.from_csv(
    "http://5.189.161.24:8080/DATA/Spon_count_floskelwolke_gutmensch.csv", 
    index_col=0)['gutmensch_relativ'].to_frame()
df.columns = ['gutmensch']

df['flchtling'] = pd.DataFrame.from_csv(
    "http://5.189.161.24:8080/DATA/Spon_count_fl%C3%BCchtling.csv", 
    index_col=0)['flchtling_relativ']

plot_df(df)
```
